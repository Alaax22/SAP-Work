```python
import matplotlib.pyplot as plt
import numpy as np

fig,ax=plt.subplots()

x=np.linspace(0,20,1000)
ax.plot(x,np.cos(x))
ax.axis('equal')
ax.annotate('local minimum',xy=(6.28,1),xytext=(10,4),arrowprops=dict(facecolor='orange',shrink=0.06))
ax.annotate('local maximum',xy=(5*np.pi,-1),xytext=(2,-4),arrowprops=dict(arrowstyle='->',facecolor='yellow'));
```

```python
import matplotlib.pyplot as plt
w=4
h=3
d=150
plt.figure(figsize=(w,h),dpi=d)
plt.axis([0,5,0,5])
x=[0,3,5]
y=[1,4,3.5]

label_x=1
label_y=4
arrow_x=3
arrow_y=4

arrow_properties=dict(facecolor="black",width=0.5,headwidth=4,shrink=0.1)

plt.annotate("maximum",xy=(arrow_x,arrow_y),xytext=(label_x,label_y),arrowprops=arrow_properties)
plt.plot(x,y)
```
